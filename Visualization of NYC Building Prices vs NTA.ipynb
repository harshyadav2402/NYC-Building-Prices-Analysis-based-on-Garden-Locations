```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv('C:/Users/Harsh Yadav/Downloads/DOF__Condominium_comparable_rental_income___Manhattan_-_FY_2010_2011.csv', header=0)
```

```python
fig = plt.figure()
ax = plt.subplot(111)

df1['MANHATTAN_CONDOMINIUM_PROPERTY_Market_Value_per_SqFt'].groupby([df1.NTA]).mean().sort_values(ascending= False).plot(kind='bar')
fig.set_size_inches(15, 10.5, forward=True)
plt.title('Avg Amount of NTP per SqFt')
ax.set_ylabel('NTP Average amount per SqFt')

plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)

df1['MANHATTAN_CONDOMINIUM_PROPERTY_Market_Value_per_SqFt'].groupby([df1.MANHATTAN_CONDOMINIUM_PROPERTY_Neighborhood]).mean().sort_values(ascending= False).plot(kind='bar')
fig.set_size_inches(15, 10.5, forward=True)
plt.title('Avg Amount of Neighbourhood per SqFt')
ax.set_ylabel('Neighbourhood Average amount per SqFt')

plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)

df1['MANHATTAN_CONDOMINIUM_PROPERTY_Est._Gross_Income'].groupby([df1.NTA]).mean().sort_values(ascending= False).plot(kind='bar')
fig.set_size_inches(15, 10.5, forward=True)
plt.title('Avg Gross Income of Property per NTA')
ax.set_ylabel('NTA')

plt.show()
```
